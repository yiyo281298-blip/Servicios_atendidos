# 1. Identificación del Problema
El objetivo es entender los datos de los Servicios Atendidos por la Inspección de Policía para poder predecir la cantidad de servicios (CANTIDAD) según el año, mes y tipo de servicio.

# 2. Recolección de Datos y 3. Preparación de Datos (ETL y EDA)
En esta sección cargamos los datos, verificamos nulos y exploramos las variables.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos
file_path = 'datos_inspeccion.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
# Verificación de nulos y tipos de datos
df.info()
print("\nValores nulos:\n", df.isnull().sum())
```

```python
# Limpieza básica: Asegurar que AÑO sea un string limpio sin comas
df['AÑO'] = df['AÑO'].astype(str).str.replace(',', '')
df.head()
```

## Análisis Exploratorio (EDA)

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['CANTIDAD'], bins=20, kde=True, color='blue')
plt.title('Distribución de la Cantidad de Servicios')
plt.xlabel('Cantidad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Preparación para el modelado
df.to_csv('datos_limpios.csv', index=False)
print('Datos limpios exportados a datos_limpios.csv')
```
